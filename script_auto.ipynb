# Overall dive-deep 

```python
# First we check which type of combination between fee categories and their name exist and their count

df[['fee_category','fee_name','balance_transaction_id']].groupby(['fee_category','fee_name']).size().reset_index(name='nb_transaction')
```

```python
# Then we check which card brand could be involved in the transactions, and which type of card it is (debit/credit)

df[['card_brand','card_funding']].groupby(['card_brand','card_funding']).size().reset_index(name='nb_transaction')
```

```python
# Then we can combine both and display count of each card brand/funding combination and their corresponding fee nale/category

df[['card_brand','card_funding','fee_category','fee_name','balance_transaction_id']].groupby(['card_brand','card_funding','fee_category','fee_name']).count().rename(columns={'balance_transaction_id': 'nb'})

```

```python
# Finally we compute fixed and variables mean fees for each combination

df[['card_brand','card_funding','fee_category','fee_name','event_type','plan_name','fixed_per_item_amount','variable_volume_amount','total_amount','tax_amount']].groupby(['event_type','card_brand','card_funding','fee_category','fee_name','plan_name'])[['fixed_per_item_amount','variable_volume_amount','total_amount','tax_amount']].sum()
```

```python
# Finally we group by the type of event and save it as csv

df.card_brand = df['card_brand'].fillna(value='unknown')
df.card_funding = df['card_funding'].fillna(value='unknown')
df.plan_name = df['plan_name'].fillna(value='unknown')

to_dl = df[['card_brand','card_funding','fee_category','fee_name','event_type','plan_name','variable_rate','fixed_per_item_amount','variable_volume_amount','total_amount','tax_amount','balance_transaction_id']].groupby(['event_type','card_brand','card_funding','fee_category','fee_name','plan_name']
        ).agg(
    fixed_fees_total=('fixed_per_item_amount', 'sum'),
    volume_variable_fees_total=('variable_volume_amount', 'sum'),
    total_fees_amount=('total_amount', 'sum'),
    fixed_fees_mean=('fixed_per_item_amount', 'mean'),
    variable_rate_mean=('variable_rate', 'mean'),
    by_transaction_occuring=('balance_transaction_id','nunique'),
    total_occuring = ('balance_transaction_id','count')
)

# to_dl_bis = df[['card_brand','card_funding','fee_category','fee_name','event_type','plan_name','fixed_per_item_amount','variable_rate','total_amount','tax_amount']].groupby(['event_type','card_brand','card_funding','fee_category','fee_name','plan_name'])[['fixed_per_item_amount','variable_rate']].mean()

to_dl.to_csv('stripe_p&l.csv')
```
